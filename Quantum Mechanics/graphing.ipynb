```python
from math import sqrt, pi
import numpy as np
def f1(x):
    return np.exp(-x)
def f2(x):
    return np.exp(-x) * x
def schrodinger(x):
    return np.exp(-x/2) * 2
Z = 1  # 氢原子
alpha = 1 / 137.036  # 精细结构常数
a = 1  # 玻尔半径（单位化）
kappa = -1  # 总角动量量子数
gamma = np.sqrt(kappa**2 - (Z * alpha)**2)  # 相对论参数

# 归一化常数
N = 1
print(gamma)

def dirac(r):
    rho = 2 * Z * r / a
    return N * (1 + gamma) * np.exp(-rho / 2) * rho**(gamma - 1)
#dirac函数的归一化常数N可以通过积分计算得到
# 归一化常数N的计算
from scipy.integrate import quad

def normalization_constant():
    integral, _ = quad(lambda r: dirac(r)**2 * r**2, 0, np.inf)
    return integral

N=1/normalization_constant()  # 归一化常数

print("dirac函数的归一化常数N:", normalization_constant())
N = 1 / normalization_constant()  # 归一化常数
print("归一化常数N:", N)

import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad

x = np.linspace(0, 10, 10000)
y1 = f1(x)
y2 = f2(x)
schrodinger_y = schrodinger(x)
dirac_y = dirac(x)
plt.plot(x, y1, label='func')
plt.plot(x,schrodinger_y, label='schrodinger')
plt.plot(x, dirac_y, label='dirac')
plt.title('Exponential Functions for Hydrogen Atom')
plt.xlabel(r'$\rho$')
plt.ylabel(r'$R(x)$')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

```python
from numpy import exp, sqrt, pi
def ground_state_energy(Z):
    return 4*(exp(-2*Z))*Z

import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
x=np.linspace(0, 10, 10000)
y1=ground_state_energy(x)

plt.plot(x, y1, label='ground_state_energy')
plt.title('Ground State Energy of Hydrogen Atom')
plt.xlabel(r'$\rho$')
plt.ylabel(r'$E_0$')
plt.ylabel('E(x)')
plt.legend()
plt.grid()
plt.show()
```
