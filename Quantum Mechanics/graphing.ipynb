```python
from math import sqrt, pi
import numpy as np
def f1(x):
    return np.exp(-x)
def f2(x):
    return np.exp(-x) * x
def schrodinger(x):
    return np.exp(-x/2) * 2
Z = 1  # 氢原子
alpha = 1 / 137.036  # 精细结构常数
a = 1  # 玻尔半径（单位化）
kappa = -1  # 总角动量量子数
gamma = np.sqrt(kappa**2 - (Z * alpha)**2)  # 相对论参数

# 归一化常数
N = 1
print(gamma)

def dirac(r):
    rho = 2 * Z * r / a
    return N * (1 + gamma) * np.exp(-rho / 2) * rho**(gamma - 1)
#dirac函数的归一化常数N可以通过积分计算得到
# 归一化常数N的计算
from scipy.integrate import quad

def normalization_constant():
    integral, _ = quad(lambda r: dirac(r)**2 * r**2, 0, np.inf)
    return integral

N=1/normalization_constant()  # 归一化常数

print("dirac函数的归一化常数N:", normalization_constant())
N = 1 / normalization_constant()  # 归一化常数
print("归一化常数N:", N)

import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad

x = np.linspace(0, 10, 10000)
y1 = f1(x)
y2 = f2(x)
schrodinger_y = schrodinger(x)
dirac_y = dirac(x)
plt.plot(x, y1, label='func')
plt.plot(x,schrodinger_y, label='schrodinger')
plt.plot(x, dirac_y, label='dirac')
plt.title('Exponential Functions for Hydrogen Atom')
plt.xlabel(r'$\rho$')
plt.ylabel(r'$R(x)$')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

```python
import plotly.graph_objects as go

# 创建交互式图表
fig = go.Figure()

# 添加 f1 曲线
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='func'))

# 添加 schrodinger 曲线
fig.add_trace(go.Scatter(x=x, y=schrodinger_y, mode='lines', name='schrodinger'))

# 添加 dirac 曲线
fig.add_trace(go.Scatter(x=x, y=dirac_y, mode='lines', name='dirac'))

# 设置图表标题和轴标签
fig.update_layout(
    title='Exponential Functions for Hydrogen Atom',
    xaxis_title='x',
    yaxis_title='f(x)',
    template='plotly_white'
)

# 显示交互式图表
fig.show()
```
